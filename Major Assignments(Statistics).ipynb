```python
import scipy.stats as stats

# Given information
sample_mean = 60
population_mean = 50
sample_std = 30
sample_size = 25
significance_level = 0.05

# Calculate the t-value
t = (sample_mean - population_mean) / (sample_std / (sample_size ** 0.5))

# Calculate the critical t-value
critical_t = stats.t.ppf(1 - significance_level/2, sample_size - 1)

# Perform the hypothesis test
if abs(t) > critical_t:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
from scipy.stats import norm

mean = 50
std_dev = 15

# Calculate the z-scores for the given values
z1 = (50 - mean) / std_dev
z2 = (70 - mean) / std_dev

# Calculate the probability using the cumulative distribution function (CDF)
prob = norm.cdf(z2) - norm.cdf(z1)

# Convert the probability to percentage
percentage = prob * 100

print("The probability that the length of time will be between 50 and 70 hours is:", percentage, "%")
```

</b>Question no: 07</br>

```python
# Probability of an email being spam
p_spam = 0.5

# Probability of detecting spam given that an email is spam
p_detection_given_spam = 0.99

# Probability of a false positive (non-spam detected as spam)
p_false_positive = 0.05

# Probability of detecting non-spam given that an email is non-spam
p_detection_given_non_spam = 1 - p_false_positive

# Applying Bayes' theorem
p_non_spam_given_detection = (p_detection_given_non_spam * (1 - p_spam)) / ((p_detection_given_non_spam * (1 - p_spam)) + (p_detection_given_spam * p_spam))

print("The probability that an email detected as spam is actually a non-spam email is:", p_non_spam_given_detection)
```
